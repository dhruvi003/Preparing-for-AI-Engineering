```python
## setting up chromadb with sentence-transformers
embed_fn = embedding_functions.SentenceTransformerEmbeddingFunction(
    model_name = "all-MiniLM-L6-v2"
)
```

```python
chroma_client = chromadb.Client()
```

```python
collection = chroma_client.create_collection(
    name = "ml_knowledge_base",
    embedding_function = embed_fn
)
```

```python
collection.add(
    documents = [d['text'] for d in DOCUMENTS],
    ids = [d['id'] for d in DOCUMENTS],
)
```

```python
collection.count()
```

```python
def retrieval(query: str, n_doc: int =3 ) -> str:
    """ Search the knowledge base and return top match chunk """
    results = collection.query(query_texts=[query], n_results=n_doc)

    docs = results['documents'][0]
    if not docs:
        return "No relevant documents were found"
    formatted = "\n\n".join(
        f'chunk {i+1}: {d} ' for i, d in enumerate(docs)
    )
    return formatted
```

```python
## Agentic RAG
```

```python
RETRIEVAL_TOOL = [{
    "type": "function",
    "function": {
        "name": "retrieve",
        "description": (
            "Search the ML knowledge base for relevant information. "
            "Call this whenever you need facts to answer a question. "
            "If the first results are not relevant enough, call again "
            "with a more specific query."
        ),
        "parameters": {
            "type": "object",
            "properties": {
                "query": {
                    "type": "string",
                    "description": "Specific search query — use keywords, not full questions"
                },
                "n_results": {
                    "type": "string", # Changed from integer to string
                    "description": "Number of chunks to retrieve (1-5, default 3)",
                    "default": 3
                }
            },
            "required": ["query"]
        }
    }
}]
```

```python
SYSTEM_PROMPT = """You are a knowledgable AI Assitant with access to ML Knowledge base
Rules:
1. ALWAYS use the retrieve tool before answering factual questions.
2. If retrieved results don't fully answer the question, retrieve again with a better query.
3. Only use information from retrieved chunks — do not add outside knowledge.
4. If the knowledge base doesn't contain the answer, say so honestly.
5. Cite which chunks supported your answer (e.g. "According to Chunk 2...").
"""
```

```python
def agentic_rag(question: str, verbose: bool = True)-> dict:
    """
      Agentic RAG Loop
      returns dictionary with: answer, retrieved_chunks, num_retrievals
    """

    messages = [
        {"role":"system", "content": SYSTEM_PROMPT},
        {"role":"user", "content": question}
    ]
    num_retrievals = 0
    chunks = []


    for step in range(4):
        res = client.chat.completions.create(
            model = model,
            messages = messages,
            tools = RETRIEVAL_TOOL,
            tool_choice= "auto",
            temperature= 0
        )

        a = res.choices[0].message
        messages.append({
            "role":"assistant",
            "content": a.content,
            "tool_calls" : [
                {
                    "id": tc.id, "type": "function",
                    "function":
                    {
                        "name": tc.function.name,
                        "arguments": tc.function.arguments
                    }
                }
                for tc in (a.tool_calls or [])
            ]
        })

        if not a.tool_calls:
            if verbose:
              print(f'answer after {num_retrievals} retrievals: ')
              print(a.content)
            return {
                "answer": a.content,
                "retrieved_chunks": chunks,
                "num_retrievals": num_retrievals,
            }

        for tc in a.tool_calls:
            args = json.loads(tc.function.arguments)
            query = args["query"]
            n = args.get("n_results", 3)
            if verbose:
                print(f"\n[retrieve] query='{query}', n={n}")

            result = retrieval(query, n_doc=n)
            chunks.append({"query": query, "result": result})
            num_retrievals += 1

            if verbose:
                print(f"[retrieve] Got {result.count('[Chunk')} chunk(s)")

            messages.append({
                "role": "tool",
                "tool_call_id": tc.id,
                "name": "retrieve",
                "content": result,
            })

    return {"answer": "Max steps reached.", "retrieved_chunks": chunks,
            "num_retrievals": num_retrievals}
```

```python
r2 = agentic_rag(
    "How does RAG reduce hallucinations, and what framework "
    "would you use to evaluate whether it's working?"
)
```

```python
# ── Cell 3: Query rewriting demonstration ─────────────
```

```python
REWRITE_SYSTEM = """You are an AI assistant with access to an ML knowledge base.

IMPORTANT: You must retrieve information before answering.
If your first retrieval returns chunks that are not directly relevant,
you MUST try again with a more specific or reworded query.
Keep trying until you find relevant information or exhaust 3 attempts."""
```

```python
def agentic_rag_with_rewrite(query: str) -> dict:
    messages = [
        {"role":"system", "content": REWRITE_SYSTEM},
        {"role":"user", "content": query}
    ]

    queries_tried = []
    for step in range(3):
        res = client.chat.completions.create(
            model = model,
            messages = messages,
            tools = RETRIEVAL_TOOL,
            tool_choice= "auto",
            temperature= 0
        )

        msg = res.choices[0].message
        messages.append(
            {
            "role": "assistant", "content": msg.content,
            "tool_calls": [{
                "id": tc.id, "type": "function",
                "function": {"name": tc.function.name, "arguments": tc.function.arguments}
            } for tc in (msg.tool_calls or [])]
        }
        )

        if not msg.tool_calls:
            print(f"\nFinal answer:\n{msg.content}")
            print(f"\nQueries tried ({len(queries_tried)}):")
            for i, q in enumerate(queries_tried, 1):
                print(f"  {i}. '{q}'")
            return {
                "answer": msg.content,
                "queries_tried": queries_tried
            }
        for tc in msg.tool_calls:
            args = json.loads(tc.function.arguments)
            query = args.get("query","")
            queries_tried.append(query)
            print(f"[attempt {len(queries_tried)}] Retrieving: '{query}'")
            # Convert n_results to an integer before passing to retrieval
            n_results = int(args.get("n_results", 3)) # Ensure conversion to int
            result = retrieval(query, n_doc=n_results)
            messages.append({
                "role": "tool", "tool_call_id": tc.id,
                "name": "retrieve", "content": result,
            })
        return {"answer": "Max steps.", "queries": queries_tried}
```

```python
result = agentic_rag_with_rewrite(
    "Tell me about the thing that helps language models "

)
```

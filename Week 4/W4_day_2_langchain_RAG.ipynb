```python
from langchain_community.document_loaders import TextLoader, PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
# from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_groq import ChatGroq
from langchain_core.output_parsers import StrOutputParser
# import chromadb
```

```python
sample_text = """
LangChain is a framework for building LLM-powered applications.
It provides abstractions called Runnables that share a standard interface.

RAG (Retrieval-Augmented Generation) improves LLM accuracy by fetching
relevant documents before generating a response. The main steps are:
1. Load and chunk documents into smaller pieces
2. Embed chunks and store in a vector database
3. At query time, retrieve the top-k most similar chunks
4. Pass chunks + question to the LLM as context

Fine-tuning adapts a pre-trained model to a specific task by training
on a smaller labelled dataset. LoRA (Low-Rank Adaptation) makes this
efficient by only updating small adapter matrices instead of all weights.

Vector databases like ChromaDB, Pinecone, and Weaviate store high-dimensional
embeddings and enable fast approximate nearest-neighbour search. They are the
storage layer that makes semantic search at scale possible.
"""
```

```python
with open("knowledge.txt", "w") as f:
    f.write(sample_text)
```

```python
loader = TextLoader("knowledge.txt")
```

```python
docs = loader.load()
```

```python
len(docs)
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size=50, chunk_overlap=10, separators='\n')
```

```python
chunks = splitter.split_documents(docs)
```

```python
embedding = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
# vectorstore = Chroma.from_documents(chunks, embedding)
a = Chroma.from_documents(chunks, embedding)
```

```python
retriever = a.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 3}  # retrieve top 3 chunks
)
```

```python
from langchain.chains import ConversationalRetrievalChain
from langchain.memory import ConversationBufferMemory

llm = ChatGroq(model="llama3-8b-8192", temperature=0)

# memory_key must match chain's input key
# return_messages=True stores as message objects (required for chat models)
memory = ConversationBufferMemory(
    memory_key="chat_history",
    return_messages=True,
    output_key="answer"
)

# Makes 2 LLM calls per turn:
# 1. Rewrite question using history into a standalone question
# 2. Retrieve docs + generate the final answer
qa_chain = ConversationalRetrievalChain.from_llm(
    llm=llm,
    retriever=retriever,
    memory=memory,
    return_source_documents=True,
)

# Turn 1 — full question
r1 = qa_chain.invoke({"question": "What is RAG and what are its main steps?"})
print("Q1:", r1["answer"])
print(f"Sources: {len(r1['source_documents'])} chunks used")

# Turn 2 — vague follow-up: memory rewrites "it" into a full question
r2 = qa_chain.invoke({"question": "What role does the vector database play in it?"})
print("\nQ2:", r2["answer"])

# Turn 3 — new topic, same session
r3 = qa_chain.invoke({"question": "How does LoRA differ from full fine-tuning?"})
print("\nQ3:", r3["answer"])
```

```python
# See exactly what memory stored after 3 turns
print("=== Memory contents ===")
for msg in memory.chat_memory.messages:
    role = "Human" if "Human" in msg.__class__.__name__ else "AI"
    print(f"\n[{role}]: {msg.content[:120]}...")

# Fresh chain with clean memory for interactive use
fresh_memory = ConversationBufferMemory(
    memory_key="chat_history",
    return_messages=True,
    output_key="answer"
)
chat = ConversationalRetrievalChain.from_llm(
    llm=llm,
    retriever=retriever,
    memory=fresh_memory,
    return_source_documents=False,
)

print("\n=== Chat loop — type 'quit' to exit ===")
while True:
    user_input = input("You: ").strip()
    if user_input.lower() in ["quit", "exit", ""]:
        break
    res = chat.invoke({"question": user_input})
    print(f"AI: {res['answer']}\n")
```

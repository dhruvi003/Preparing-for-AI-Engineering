```python
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
```

```python
lllm = ChatGroq(
    api_key = "...",
    model = "Llama-3.3-70B-Versatile",
    temperature=0
)
```

```python
prompt = ChatPromptTemplate.from_messages([
    ("system","You are a concise AI tutor, explain concepts in 3 bullet points"),
    ("human","Explain {topic} like you are teaching a 5 year old")
])
```

```python
parser = StrOutputParser()
```

```python
chain = prompt | lllm | prompt
```

```python
res = chain.invoke({"topic":"Langchain"})
```

```python
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
docs = [
    "RAG stands for Retrieval-Augmented Generation. It fetches relevant docs before generating.",
    "ChromaDB is an open-source vector database that stores embeddings for semantic search.",
    "LoRA is a parameter-efficient fine-tuning method that adds low-rank matrices to model weights.",
    "The ReAct pattern combines reasoning (Thought) and acting (Action) in LLM agent loops.",
]
```

```python
embeddings = HuggingFaceEmbeddings(model_name='all-MiniLM-L6-v2')
```

```python
vector_store = Chroma.from_texts(docs, embeddings)
```

```python
retriever = vector_store.as_retriever(search_kwargs={'k':2})
```

```python
rag_prompt = ChatPromptTemplate.from_messages([
    ("system", "Answer using ONLY the context below. If unsure, say so.\n\nContext:\n{context}"),
    ("human", "{question}"),
])
```

```python
def format_docs(docs):
    return "\n\n".join(d.page_content for d in docs)
```

```python
rag_chain = (
    {"context":retriever | format_docs, "question": RunnablePassthrough()}
    | rag_prompt
    | lllm
    | parser
)
```

```python
answer = rag_chain.invoke("What is RAG and how does it work?")
print(answer)
```

```python
answer2 = rag_chain.invoke("How does LoRA fine-tuning work?")
print(answer2)
```
